... ---> need to curate the data: a simple way to do it is to set to __lower case__ all the entries:

```python
%%time

publications['n_authors'] = np.zeros(publications.shape[0],dtype=int)
for row in publications.itertuples():
    
    if type(row.Authors)==str:
        
        author1_list = row.Authors.split(',')
        author1_list = [p.strip() for p in author1_list]
    #     print(len(author1_list))

        publications.loc[row[0],'n_authors'] = int(len(author1_list))
```

```python
# plt.loglog(publications.n_authors.value_counts(),marker='.',linestyle='',)

# plt.xlabel('number of authors / publication',size=16)
# plt.ylabel('number of publications',size=16);
# # plt.savefig('./plots_for_paper/authors_per_publication.pdf')
```
